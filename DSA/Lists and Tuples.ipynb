### Challenge - Even Numbers
Given a list of random numbers , filter out all the Even numbers from it.
Do it in one line of code

Input : [24 , 27 , 21 , 46 , 48 , 6 , 25 , 38 , 10 , 40]


Output : [24 , 46 , 48, 6 , 38 , 10 , 40 ]

```python
a = [24 , 27 , 21 , 46 , 48 , 6 , 25 , 38 , 10 , 40]
b = []

for x in a :
    if x % 2 == 0 :
        b.append(x)
```

```python
#In Single Line
b= [x for x in a if x % 2 == 0]
```

```python
b
```

## Challenge - Check if Sorted
Given a list of numbers , chech if it sorted (in increasing order)

Input : [1 , 4 , 10 , 12 , 14 , 18 , 20]

```python
a = [1 , 4 , 10 , 12 , 14 , 18  , 20]
```

```python
ans ="Sorted"
for i in range(len(a)-1):
    if a[i] < a[i+1] :
        # Correct Order
        continue
    else:
        ans ="Not Sorted"
        break

print(ans)
```

## Functions - args
Article Link - https://www.scaler.com/topics/python/args-and-kwargs-in-python/

Challenge - write a function which returns the product of all its arguments


The user should be able to call the function like 

  Product(2 , 3 , 9 , 1)   --------> #54

  
  Product(4 , 2) ------------------> #8

  
The function should be able to accept any number of arguments

```python
def product(*nums):
    print(nums , type(nums))
```

```python
product(10,20,30,4,2,8,10,12)
```

```python
def speak(*words):
    print(words ,type(words))
```

```python
speak("Hello" ,'my' , 'name' , 'is' , 'Teja')
```

```python
def products(*args):
    p=1
    for number in args:
        p = p * number
    return p
```

```python
products(2,3,9,1)
```
